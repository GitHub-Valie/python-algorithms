# Load, prepare and visualize data in Python

This program loads a csv file, prepares the data and plots some of it using the Pandas and Matplotlib packages.

```python
import pandas
import matplotlib.pyplot as plt
```

## Loading data

Next, we load the csv file into a Pandas DataFrame, and we assign column names to our dataframe 

```python
df = pandas.read_csv(
    'data\BTCUSDT.csv', 
    names = [
        'Open time', 
        'Open', 
        'High', 
        'Low', 
        'Close', 
        'Volume', 
        'Close time', 
        'Quote asset volume', 
        'Number of trades', 
        'Taker BUY base asset volume', 
        'Taker BUY quote asset volume', 
        'Ignore'
    ]
)
```

## Preparing data

For plotting, we will be using the `Open time` as our x axis, so we set this column as our dataframe index.

We then clean our dataframe by dropping the columns we won't be using.

```python
df.index = pandas.to_datetime(
    df['Open time'], 
    unit='s'
)

df = df.drop(
    columns = [
        'Open time',
        'Close time',
        'Quote asset volume', 
        'Taker BUY base asset volume', 
        'Taker BUY quote asset volume', 
        'Ignore'
    ]
)
```

## Explore data

Let's take a look at our data now that it's prepared

```python
print(df.head())
```

Now, let's take a look at the descriptive stats of our data

```python
print(df.describe())
```

## Visualize data

Finally, let's plot the closing prices of BTCUSDT 1 minute

```python
df.plot(y=['Close'], title = 'BTCUSDT, 1 MINUTE')
plt.show()
```
